## WEEK-3: Basic Plots for Data Exploration (Iris Dataset)

### 1. Load the Iris Dataset

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Create a DataFrame from the dataset
# The feature names are available in iris.feature_names
# The target variable (species) is in iris.target, and target names are in iris.target_names
df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df_iris['species'] = iris.target # Add the target variable

# Display the first 5 rows of the DataFrame to inspect the data
print("First 5 rows of the Iris DataFrame:")
display(df_iris.head())

# Display basic information about the DataFrame
print("\nDataFrame Info:")
df_iris.info()
```

### 2. Generate Box Plots for Each of the Four Predictors

```python
# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Create a figure with subplots to display all four box plots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))
fig.suptitle('Box Plots of Iris Predictor Variables', fontsize=16)

# List of predictor columns
predictors = df_iris.columns[:-1] # Exclude the 'species' column

# Iterate through each predictor and create a box plot
for i, feature in enumerate(predictors):
    row = i // 2
    col = i % 2
    sns.boxplot(y=df_iris[feature], ax=axes[row, col])
    axes[row, col].set_title(f'Box Plot of {feature.replace(" (cm)", "")}', fontsize=12)
    axes[row, col].set_ylabel(feature, fontsize=10)

# Adjust layout to prevent overlapping titles and labels
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

### 3. Generate Box Plot for a Specific Feature (e.g., 'petal length (cm)')

```python
# Specify the feature for the box plot
specific_feature = 'petal length (cm)'

# Create the box plot for the specific feature
plt.figure(figsize=(8, 6))
sns.boxplot(y=df_iris[specific_feature])
plt.title(f'Box Plot of {specific_feature.replace(" (cm)", "")}', fontsize=14)
plt.ylabel(specific_feature, fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### 4. Generate Histogram for a Specific Feature (e.g., 'sepal width (cm)')

```python
# Specify the feature for the histogram
specific_feature_hist = 'sepal width (cm)'

# Create the histogram for the specific feature
plt.figure(figsize=(10, 7))
sns.histplot(df_iris[specific_feature_hist], kde=True, bins=20, color='skyblue')
plt.title(f'Histogram of {specific_feature_hist.replace(" (cm)", "")}', fontsize=14)
plt.xlabel(specific_feature_hist, fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### 5. Generate Scatter Plot of Petal Length vs. Sepal Length

```python
# Create a scatter plot
plt.figure(figsize=(10, 7))
# 'hue' parameter can be used to color points based on the 'species' column,
# which helps visualize relationships within different species.
sns.scatterplot(x='sepal length (cm)', y='petal length (cm)', hue='species', data=df_iris, palette='viridis', s=100, alpha=0.8)
plt.title('Scatter Plot of Petal Length vs. Sepal Length by Species', fontsize=14)
plt.xlabel('Sepal Length (cm)', fontsize=12)
plt.ylabel('Petal Length (cm)', fontsize=12)
plt.legend(title='Species', labels=iris.target_names) # Use target names for legend
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
