```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
plt.figure(figsize=(10,6))
df.boxplot()
plt.title("Boxplot of All Iris Features", fontsize=14)
plt.xticks(rotation=30)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.boxplot(df["petal length (cm)"])
plt.title("Boxplot of Petal Length", fontsize=14)
plt.ylabel("Petal Length (cm)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["sepal length (cm)"], bins=10)
plt.title("Histogram of Sepal Length", fontsize=14)
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df["sepal length (cm)"], df["petal length (cm)"])
plt.title("Petal Length vs Sepal Length", fontsize=14)
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.boxplot(df["sepal length (cm)"])
plt.title("Sepal Length")

plt.subplot(2,2,2)
plt.boxplot(df["sepal width (cm)"])
plt.title("Sepal Width")

plt.subplot(2,2,3)
plt.boxplot(df["petal length (cm)"])
plt.title("Petal Length")

plt.subplot(2,2,4)
plt.boxplot(df["petal width (cm)"])
plt.title("Petal Width")

plt.tight_layout()
plt.show()
```
