## WEEK-2: Basic Statistical Measures

This section covers fundamental statistical concepts including measures of central tendency, data spread, and position of data values. We will use a sample dataset to demonstrate each measure.

```python
# Import necessary libraries
import numpy as np
import pandas as pd # For potential future use, though not strictly needed for these basics

# 1. Generate a sample dataset
# Let's create a simple array of numbers to work with
data = np.array([10, 12, 12, 15, 16, 18, 20, 20, 20, 22, 25, 30])
print("Sample Data:", data)
```

### 1. Measures of Central Tendency

Measures of central tendency help us find the 'center' or 'average' of a dataset.

*   **Mean**: The sum of all values divided by the number of values. It's the most common average.
*   **Median**: The middle value in a dataset that is ordered from least to greatest. If there's an even number of observations, it's the average of the two middle numbers.
*   **Mode**: The value that appears most frequently in a dataset. A dataset can have one mode (unimodal), multiple modes (multimodal), or no mode at all.

```python
# Calculate Mean
mean_value = np.mean(data)
print(f"Mean: {mean_value:.2f}")

# Calculate Median
median_value = np.median(data)
print(f"Median: {median_value:.2f}")

# Calculate Mode
# NumPy doesn't have a direct mode function, so we'll use a manual approach or scipy.stats.mode
# For simplicity, let's do it manually for now, or use pandas value_counts()
from collections import Counter

counts = Counter(data)
max_count = max(counts.values())
mode_values = [key for key, value in counts.items() if value == max_count]
print(f"Mode: {mode_values}")
```

### 2. Measures of Data Spread (Dispersion)

Measures of data spread (or dispersion) describe how spread out or scattered the data points are. They tell us about the variability within the dataset.

*   **Variance**: The average of the squared differences from the Mean. It measures how far each number in the set is from the mean and therefore from every other number in the set.
*   **Standard Deviation**: The square root of the variance. It's the most common measure of spread and indicates the typical distance between data points and the mean. A low standard deviation means data points are generally close to the mean, while a high standard deviation means data points are spread out over a wider range of values.

```python
import numpy as np

# Calculate Variance
# ddof=0 for population variance, ddof=1 for sample variance (default in pandas, numpy default is 0)
variance = np.var(data, ddof=0) # Population variance
print(f"Variance (population): {variance:.2f}")

# Calculate Standard Deviation
std_dev = np.std(data, ddof=0) # Population standard deviation
print(f"Standard Deviation (population): {std_dev:.2f}")

# If you need sample variance/std dev (e.g., when working with a sample of a larger population)
sample_var = np.var(data, ddof=1)
sample_std = np.std(data, ddof=1)
print(f"Variance (sample): {sample_var:.2f}")
print(f"Standard Deviation (sample): {sample_std:.2f}")
```

### 3. Position of Different Data Values (Quartiles, IQR)

These measures help us understand the position of specific data points within the dataset and the spread of the middle 50% of the data.

*   **Quartiles**: Divide a dataset into four equal parts. They are:
    *   **Q1 (First Quartile)**: 25th percentile, separates the lowest 25% of data from the highest 75%.
    *   **Q2 (Second Quartile)**: 50th percentile, which is also the Median.
    *   **Q3 (Third Quartile)**: 75th percentile, separates the lowest 75% of data from the highest 25%.
*   **Inter-Quartile Range (IQR)**: The range between the first and third quartiles (Q3 - Q1). It represents the middle 50% of the data and is a robust measure of statistical dispersion, less sensitive to outliers than the standard deviation.

```python
import numpy as np
from collections import Counter

# Calculate Mean
mean = np.mean(data)
print(f"Mean: {mean:.2f}")

# Calculate Median
median = np.median(data)
print(f"Median: {median:.2f}")

# Calculate Mode
counts = Counter(data)
max_count = max(counts.values())
mode = [key for key, value in counts.items() if value == max_count]
print(f"Mode: {mode}")
```

```python
# Calculate Quartiles
# np.percentile takes the data and the desired percentile(s)
q1 = np.percentile(data, 25)
q2 = np.percentile(data, 50) # This should be the same as the median
q3 = np.percentile(data, 75)

print(f"First Quartile (Q1): {q1:.2f}")
print(f"Second Quartile (Q2 - Median): {q2:.2f}")
print(f"Third Quartile (Q3): {q3:.2f}")

# Calculate Inter-Quartile Range (IQR)
iqr = q3 - q1
print(f"Inter-Quartile Range (IQR): {iqr:.2f}")
```
