## Feature Construction Techniques

This section demonstrates various feature construction techniques:
1.  **Dummy Coding** for nominal categorical variables.
2.  **Ordinal Encoding** for ordinal categorical variables.
3.  **Transforming numeric (continuous) features** to categorical features.

```python
import pandas as pd
import numpy as np

# Create sample data for demonstration
data = {
    'ID': range(1, 11),
    'Color': ['Red', 'Blue', 'Green', 'Red', 'Blue', 'Green', 'Red', 'Blue', 'Green', 'Red'],
    'Size': ['Small', 'Medium', 'Large', 'Small', 'Medium', 'Small', 'Large', 'Medium', 'Small', 'Large'],
    'Age': [25, 30, 45, 22, 38, 50, 28, 35, 42, 55],
    'City': ['NY', 'LA', 'Chi', 'NY', 'LA', 'Chi', 'NY', 'LA', 'Chi', 'NY'] # Another nominal for dummy coding
}
df = pd.DataFrame(data)

print("Original DataFrame:")
display(df.head())
```

### 1. Dummy Coding Nominal Categorical Variables

Dummy coding converts categorical variables into a numerical format that can be used by machine learning algorithms. For each unique category, a new binary column (0 or 1) is created. This is often used for nominal variables where there is no inherent order.

We will apply dummy coding to the 'Color' and 'City' columns.

```python
# Perform dummy coding on 'Color' and 'City' columns
df_encoded_nominal = pd.get_dummies(df, columns=['Color', 'City'], prefix=['Color', 'City'])

print("DataFrame after Dummy Coding (Nominal Variables):")
display(df_encoded_nominal.head())
```

### 2. Encoding Ordinal Categorical Variables

Ordinal encoding is used for categorical variables that have an intrinsic order or ranking. Each category is assigned an integer value based on its order.

We will encode the 'Size' column, which has an order: 'Small' < 'Medium' < 'Large'.

```python
# Define the order for the 'Size' variable
size_order = ['Small', 'Medium', 'Large']

# Create a mapping dictionary
size_mapping = {size: i for i, size in enumerate(size_order)}

# Apply the ordinal encoding to the 'Size' column
df_encoded_nominal['Size_Encoded'] = df_encoded_nominal['Size'].map(size_mapping)

print("DataFrame after Ordinal Encoding (Ordinal Variables):")
display(df_encoded_nominal.head())

# Drop the original 'Size' column if no longer needed
df_final = df_encoded_nominal.drop(columns=['Size'])
print("DataFrame after dropping original 'Size' column:")
display(df_final.head())
```

### 3. Transforming Numeric (Continuous) Features to Categorical Features

This technique, often called **binning** or **discretization**, converts continuous numerical data into discrete categories (bins). This can be useful for reducing noise, handling non-linear relationships, or preparing data for models that prefer categorical inputs.

We will transform the 'Age' column into age groups (e.g., 'Young', 'Middle-aged', 'Senior').

```python
# Define bin edges and labels for 'Age'
# For example, ages 0-29 as 'Young', 30-49 as 'Middle-aged', 50+ as 'Senior'
bins = [0, 29, 49, np.inf]
labels = ['Young', 'Middle-aged', 'Senior']

# Use pd.cut to create new categorical 'Age_Group' feature
df_final['Age_Group'] = pd.cut(df_final['Age'], bins=bins, labels=labels, right=True)

print("DataFrame after Transforming Numeric 'Age' to Categorical 'Age_Group':")
display(df_final.head())

# You can also apply dummy coding to the new 'Age_Group' if desired for certain models
df_final_with_age_dummies = pd.get_dummies(df_final, columns=['Age_Group'], prefix='Age')
print("DataFrame after Dummy Coding 'Age_Group':")
display(df_final_with_age_dummies.head())
```
