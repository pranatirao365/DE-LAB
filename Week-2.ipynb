```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("auto-mpg.csv", na_values="?")

# Convert horsepower to numeric
df["horsepower"] = pd.to_numeric(df["horsepower"], errors="coerce")

# Check missing values
df.isnull().sum()
```

```python
#Measures of Central Tendency
```

```python
#Average
mean_mpg = df["mpg"].mean()
mean_mpg
```

```python
#Median(To know the middle value of the dataset.)
median_mpg = df["mpg"].median()
median_mpg
```

```python
#Mode (Most Frequent Value)
mode_mpg = df["mpg"].mode()
mode_mpg
```

```python
#Measure of Data Spread.Range = Max âˆ’ Min
range_mpg = df["mpg"].max() - df["mpg"].min()
range_mpg
```

```python
#Dispersion of Data
variance_mpg = df["mpg"].var()
variance_mpg
```

```python
std_mpg = df["mpg"].std()
std_mpg
```

```python
#Position of Data Values
Q1 = df["mpg"].quantile(0.25)
Q2 = df["mpg"].quantile(0.50)   # Median
Q3 = df["mpg"].quantile(0.75)

Q1, Q2, Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
plt.hist(df["mpg"], bins=10)
plt.title("Distribution of MPG")
plt.xlabel("MPG")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.boxplot(df["mpg"])
plt.title("Boxplot of MPG")
plt.show()
```
